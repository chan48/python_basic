```python
class A(object) :
    print("A")
    
    def __getattribute__(self,name) :
        raise Exception("get")
    
    def __setattr__(self,name,value) :
        raise Exception
        
    
    def __setitem__(self,name,value) :
        raise Exception
        
setattr(A,"name","aa")
print(A.__dict__)
getattr(A,'__setattr__')(A(),'aaa','aaa')
```

```python
from multipledispatch import dispatch
@dispatch(int, int)
def add(x, y):
    return x + y    

@dispatch(object, object)
def add(x, y):
    return "%s + %s" % (x, y)

print(add(5,5))
print(add("a","b"))
```
