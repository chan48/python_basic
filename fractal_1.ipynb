```python
%matplotlib inline

import random
import numpy as np
import matplotlib.pyplot as plt

v_16 = 1.6
v_085 = 0.85
v_004 = 0.04
v_015 = 0.15
v_016 = 0.16
v_02  = 0.02
v_022 = 0.22
v_023 = 0.23
v_024 = 0.24
v_026 = 0.26
v_028 = 0.28
v_044 = 0.44

def transformation(num,x,y) :
   
    if num == 0 :
        x1 = v_085*x + v_004*y
        y1 = -1*v_004*x + v_085*y + v_16
    elif num == 1 :
        x1 = v_02*x + -1*v_026*y
        y1 = v_023*x + v_022*y + v_16
    elif num == 2 :
        x1 = -1*v_015*x + v_028*y
        y1 = v_026*x + v_024*y + v_044
    else :
        x1 = 0
        y1 = v_016*y 
    
    return x1,y1

def get_index(probability) :
    r = np.random.random()
    c_probability= 0
    sum_probability = []
    
    for p in probability :
        c_probability += p
        sum_probability.append(c_probability)
        
    for item, sp in enumerate(sum_probability) :
        if r <= sp :
            return item
    
    return len(probability) -1

def transform(x,y) :
    probability = [0.85,0.07,0.07,0.01]
    tindex = int(get_index(probability))
    
    x1,y1 = transformation(tindex,x,y)
    return x1,y1

def draw_fern(n) :
    x=[0]
    y=[0]
    x1,y1 = 1,1
    for i in range(n) :
        x1,y1 = transform(x1,y1)
        x.append(x1)
        y.append(y1)
    print len(x),len(y)
    return x, y

if __name__  ==  '__main__' :
    n= int(input('Enter the number of points in the Fern :'))
    x,y = draw_fern(n)
    
    plt.plot(x,y, 'ro')
    plt.title('Fern with {0} points'.format(n))
    plt.show()
```
